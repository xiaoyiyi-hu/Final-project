```python
def cv_TF(row):
    if row.trial_completed_TF==False:
        return False
    elif (row.payment_period==True) and (row.payment_period==0):
        return False
    else:
        return True
```

```python
df3.loc[:,'cv_TF'] = df3.apply(cv_TF,axis=1)
```

```python
ab_df=ab_df.loc[ab_df.plan_type.isin(['base_uae_14_day_trial','low_uae_no_trial','high_uae_14_day_trial'])]
ab_df.head()
```

```python
ab_table = ab_df.pivot_table(values='cv_TF', index='plan_type', aggfunc=np.sum)
# add additional columns to the pivot table
ab_table['total'] = ab_df.pivot_table(values='cv_TF', index='plan_type', aggfunc=lambda x: len(x))
ab_table['rate'] = ab_df.pivot_table(values='cv_TF', index='plan_type')
ab_table.head()
```

```python
#treatment:high_uae_14_day_trial
#Control:base_uae_14_day_trial
A_converted = ab_table ['cv_TF'][0]
N_A = ab_table ['total'][0]
p_A = ab_table ['rate'][0]
B_converted = ab_table ['cv_TF'][1]
N_B = ab_table ['total'][1]
p_B = ab_table ['rate'][1]
```

```python
pooled_prob = (p_A*N_A+p_B*N_B)/(N_A + N_B)
z=(p_B - p_A)/(pooled_prob*(1-pooled_prob)*(1/N_A +1/N_B))**0.5
z
```

```python
#treatment:low_uae_no_trial
#Control:base_uae_14_day_trial
B_converted = ab_table ['cv_TF'][2]
N_B = ab_table ['total'][2]
p_B = ab_table ['rate'][2]
```

# Churn Modeling

1. data preprocessing

```python
#For the dataset "Engagement", calculate the daily data for each column.
Egm_data=pd.read_pickle("engagement.dms")
Egm_data.head()
```

```python
df2['current_sub_TF'].value_counts()
```

```python
Egm_data = Egm_data.groupby('subid').agg({'date': ['min', 'max'], 'app_opens':  'sum',
                                           'cust_service_mssgs' : 'sum', 'num_videos_rated': 'sum',
                                           'num_videos_completed': 'sum', 'num_videos_more_than_30_seconds': 'sum',
                                           'num_series_started': 'sum','payment_period':'max'}).reset_index()
Egm_data.columns = ['subid','day_min', 'day_max',  
                     'app_opens', 'cust_service_mssgs', 'num_videos_rated', 'num_videos_completed',
                    'num_videos_more_than_30_seconds', 'num_series_started','payment_period_max']
Egm_data['days']=(Egm_data['day_max']-Egm_data['day_min']).dt.days+1
Egm_data
```

```python
column_list=[ 'app_opens', 'cust_service_mssgs', 'num_videos_rated', 'num_videos_completed',
                    'num_videos_more_than_30_seconds', 'num_series_started']
for col in column_list:
    Egm_data[col]=Egm_data[col]/Egm_data['days']

Egm_data.head()
```

```python
Egm_data=Egm_data.drop(columns=['day_min', 'day_max'])
Egm_data.head()
```

```python
df1.columns
```

```python
#Merge 3 datasets:subscribers,customer_service_rep and engagement.
eng_csr=pd.merge(Egm_data,df2)
eng_csr.head()
```

```python
sec=pd.merge(df1,eng_csr)
sec.head()
```

```python
#Choose features for the model
sec_mdl=sec[['package_type','preferred_genre', 'intended_use','male_TF','attribution_technical','plan_type','op_sys','app_opens', 'cust_service_mssgs', 'num_videos_rated',
       'num_videos_completed', 'num_videos_more_than_30_seconds',
       'num_series_started', 'revenue_net','current_sub_TF']]
```

```python
sec_mdl.head()
```

```python
sec_mdl['revenue_net'].mean()
```

```python
sec_mdl['preferred_genre']=sec_mdl['preferred_genre'].fillna('Unknown')
sec_mdl['package_type']=sec_mdl['package_type'].fillna('Unknown')
sec_mdl['intended_use']=sec_mdl['intended_use'].fillna('Unknown')
sec_mdl['revenue_net']=sec_mdl['revenue_net'].fillna(1.71)
```

```python
# Change categorical variables into binary variables
sec_mdl.loc[sec_mdl['current_sub_TF']==True,'current_sub_TF']=1
sec_mdl.loc[sec_mdl['current_sub_TF']==False,'current_sub_TF']=0
```

```python
sec_mdl.columns
```

```python
# divid features into catergory and numerical
f_cat=['package_type', 'preferred_genre', 'intended_use', 'male_TF',
       'attribution_technical', 'plan_type', 'op_sys']
f_num=['app_opens',
       'cust_service_mssgs', 'num_videos_rated', 'num_videos_completed',
       'num_videos_more_than_30_seconds', 'num_series_started', 'revenue_net']
for i in f_cat:
    sec_mdl[i].astype('category')

# convert catergory features into dummy variables
def change_to_dummies(df):
    dm_df= pd.DataFrame()
    for i in f_cat:
        dm_df=pd.concat([dm_df,pd.get_dummies(df[i],prefix=i, drop_first=True)],axis=1)
    for j in f_num:
        dm_df[j]=df[j]
    dm_df=pd.concat([dm_df, df['current_sub_TF']], axis=1)
    return dm_df
```

```python
#Split data into training (70%) and test (30%) data set
from sklearn.model_selection import train_test_split

from matplotlib.colors import ListedColormap

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import validation_curve

from sklearn.model_selection import GridSearchCV
dm_mdl=change_to_dummies(sec_mdl)
X=dm_mdl.drop(['current_sub_TF'], axis=1)
y=dm_mdl['current_sub_TF']
X_train, X_test, y_train, y_test=train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
X_train.columns
```

2.Logistic Model

```python
from sklearn.metrics import auc, classification_report, f1_score, confusion_matrix, accuracy_score, roc_auc_score, roc_curve
clf = LogisticRegression(C=100).fit(X_train,y_train)
y_prob_predicted = clf.predict(X_test)
print('Accuracy of Logistic regression classifier on training set: {:.5f}'
     .format(clf.score(X_train, y_train)))

print('Accuracy of Logistic regression classifier on test set: {:.5f}'
     .format(clf.score(X_test, y_test)))
print(classification_report(y_test,y_prob_predicted))
```

```python
importance = pd.DataFrame(clf.coef_[0], index = X_train.columns, columns=['coefficents'])
importance.sort_values(by=['coefficents'])
```

2.Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
clf2 = DecisionTreeClassifier(max_depth = 3).fit(X_train, y_train)

print('Accuracy of Decision Tree classifier on training set: {:.8f}'
     .format(clf2.score(X_train, y_train)))
print('Accuracy of Decision Tree classifier on test set: {:.8f}'
     .format(clf2.score(X_test, y_test)))
```

```python
importance1 = pd.DataFrame(clf2.feature_importances_, index = X_train.columns, columns=['feature_importances_'])
importance1.sort_values(by=['feature_importances_'])
importance1.to_csv('coe_dt.csv')
```

3. Random Forest

```python
from sklearn.ensemble import RandomForestClassifier


clf3 = RandomForestClassifier(max_depth = 7,n_estimators=20).fit(X_train, y_train)
print('Accuracy of Random Forest classifier on training set: {:.8f}'
     .format(clf3.score(X_train, y_train)))
print('Accuracy of Random Forest classifier on test set: {:.8f}'
     .format(clf3.score(X_test, y_test)))
```

```python
importance2 = pd.DataFrame(clf3.feature_importances_, index = X_train.columns, columns=['feature_importances_'])
importance2.sort_values(by=['feature_importances_'])
importance2.to_csv('co_rf.csv')
```

```python
actual=pd.DataFrame(y_test)
actual.head(10)
actual.shape
```

```python
table_tst=pre.join(actual.reset_index())
table_tst.head(10)
```

```python
r_sec=pd.DataFrame(cac_set[['subid','attribution_technical','monthly_price','month','year','CAC']]).reset_index()
df_churn=pd.merge(table_tst,r_sec)
df_churn.head()

```

```python
# Calculate CLV for each id
# Since we set the annual discount rate as 10%, and users have to pay 3 times a year(every 4 months)
#Formula: payment[i]*(1+r)/(1+r-prb(churn))-cac
r= 0.1/3
df_churn['CLV']=df_churn['monthly_price']*(1+r)/(1+r-df_churn[0])-df_churn['CAC']
df_churn.head()
```
